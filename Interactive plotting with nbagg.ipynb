This is an alternative demonstration of interactive plotting in IPython. Here we use the new matplotlib nbagg backend which allows us to zoom and pan freely. Note how we don't create a new figure every time but just replace the data. This is much faster. Unfortunately doing this with Basemap is somewhat complicated. 

```python
from IPython.html.widgets import interact, fixed, FloatTextWidget, fixed
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pltsin(ax, f=3):
    x = np.linspace(0,1,100)
    y = np.sin(2*np.pi*x*f)
    if len(ax.lines):
        ax.lines[0].set_xdata(x)
        ax.lines[0].set_ydata(y)
    else:
        ax.plot(x,y)
    fig.canvas.draw()
```

```python
fig,ax = plt.subplots(1,1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(0,1)
ax.set_ylim(-1,1)
pltsin(ax)
```

```python
interact(pltsin, ax=fixed(ax), f=(1, 10, 0.1))
```
