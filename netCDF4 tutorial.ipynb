> *Question: What are the problems with the CSV format?*
> 
> * *We don't know what the dimensions are (can anyone even remember this from two days ago?),*
> * *We don't what the units are,*
> * *We can only store two dimensional data,*
> * *Not very efficient are storing lots of data.*

NetCDF is a data format which solves all of these problems. With NetCDF:

* Every variable is labelled so that you know what the data represents;
* You can store metadata, like the units that the data is represented in, and how missing values are represented;
* Data can have any number of dimensions. E.g You could store the temperature and moisture content of soil at different depths, across a grid of latitude and longitude, over time;
* It compresses data.

If you share and publish your data in the NetCDF format then it will contain all of the context needed to make sense of the data, even if they don't have any other external information. NetCDF files come with context about the data they represent, and context is everything! Compare this to the CSV file above, which if you found after a year without any other information, you'd have no idea what data it contained.

# Reading NetCDF Data

The first thing to do if we want to read in NetCDF data in Python is to import the netCDF4 module:

Lets first download a netCDF file to experiment with. The following file taken from http://www.metoffice.gov.uk/hadobs/hadisst/data/download.html contains the sea surface temperature for the last month.

````python
import os.path
if not(os.path.isfile('HadISST1_SST_update.nc.gz')):
    !wget http://www.metoffice.gov.uk/hadobs/hadisst/data/HadISST1_SST_update.nc.gz
````

````python
import netCDF4
````

Recall that we can always ask Python for help with anything, so if you ever need to find out more about using NetCDF with Python, just run:

````python
help(netCDF4)
````

Alternativly we can use the build in help system in IPython. This has the advantage that the help is displayed in an external pager.

````python
netCDF4?
````

The help for netCDF4 comes with a complete tutorial, which you can also view online: http://unidata.github.io/netcdf4-python/.

To read in a NetCDF file, we ask the NetCDF library to open a *Dataset*, which is the term used for a collection of variables, which can all have different units (temperatures, concentrations, etc), and be grouped into combinations of different dimensions (over time, lat & long, depth, etc).

````python
dataset = netCDF4.Dataset("HadISST1_SST_update.nc.gz", mode="r")
````

Recall that an advantages of NetCDF is that they give the data context with additional metadata. We can see some of that straight away:

````python
print dataset
````

````python
print(dataset.Title)
````

What we're really interested in though is the raw numbers. Let's see what variables this NetCDF file contains:

````python
print dataset.variables
````

Or more cleanly:
    

````python
print dataset.variables.keys()
````

A variable in NetCDF is some quantity, like a temperature, or measure of inflammation (as in the Python lesson). A variable doesn't just contain a single value, but instead contains a multi-dimensional array of values with specified dimensions. For examples, a `temperature` variable might be stored as a 2 dimensional array, where one dimension is latitudes, and one dimension is longitudes. The `temperature` variable would hold a value for each combination of latitude and longitude points. If you were measuring moisture in the soil at a single location, you might have a `moisture` variable, where one dimension was soil depth, and another dimension was time.

The `sst` variable in our NetCDF file stores the sea surface temperature. We might not have known that, but fortunately the metadata in the NetCDF file tells us:

````python
print dataset.variables["sst"]
````

The meta data also tells us other useful information like the units that the temperature is stored in (Celcius), and what the different dimensions in the 3 dimensional array represent (time, latitude, longitude).

To get a feeling of what is in this dataset lest try to inspect the data at a single point in time. We use the standard array indexing to read data into a numpy array. Here we select the first data point in time and all points along the other dimensions.

````python
sst0 = dataset.variables["sst"][0,:,:]
````

Again we can use IPython's build in inspection to learn about the data. In this case the data is a numpy masked array.

````python
sst0?
````

Since this is a numpy array we can use the numpy methods to learn more about the data.

````python
print sst0.shape
print sst0.min()
print sst0.max()
````

Let's see what this data actually looks like. We can plot a 2 dimensional array as an image in Python using the `imshow` function in matplotlib. Let's plot the sea surface temperature at the first time step in our data:

````python
%matplotlib inline
````

````python
import matplotlib.pyplot as plt
plt.imshow(sst0)
plt.colorbar()
plt.tight_layout()
````

This looks rather strange with only 2 different colours. Recall from above that the minimum temperature is -1000 c. It looks like this is used as a fill value for missing values. Lets rescale to omit that.

````python
sst0min = sst0[sst0>-1000].min()
sst0max = sst0.max()
````

````python
import matplotlib.pyplot as plt
plt.imshow(sst0, vmax=sst0max, vmin=sst0min)
plt.colorbar()
````

The units on the axis of the plot are rather unconventional. That's naturally because we didn't tell imshow which limits to use. Fortunately netCFD stores the axis as well as the main data. Remember from above that sst is a time by longitude by latitude array. We can read the values from the corresponding variables.

````python
longitude = dataset.variables['longitude'][:]
latitude = dataset.variables['latitude'][:]
````

````python
import matplotlib.pyplot as plt
plt.imshow(sst0, vmax=sst0max, vmin=sst0min, extent=[longitude.min(),longitude.max(),latitude.min(),latitude.max()])
plt.colorbar()
````

#Writing data to netCDF files

First we create a new netCDF file. Several different subtypes of netCDF exists. Here we chose to create a file in netCDF4 format.

````python
rootgrp = netCDF4.Dataset('test.nc', 'w', format='NETCDF4')
````

As you may have noticed in the sea surface temperature example above netCDF variables have dimensions attaced to them. In order to store data in the netCDF file we must first create the matching dimensions. In this example we will create some artificial data on a latitude longitude coordinate system similar to the one we saw above. To make it a little more interesting we will add another dimension. 

````python
level = rootgrp.createDimension('level', None)
time = rootgrp.createDimension('time', None)
lat = rootgrp.createDimension('lat', 73)
lon = rootgrp.createDimension('lon', 144)
````

netCDF have two diffrent types of dimensions limited and unlimited. The limited dimensions have a fixed size while the unlimited ones allows appending of a unknown number of records. In our example we mesure data on a fixed latitude/longitude grid while we mesure data at an abitrary number of points in time. 

````python
print(lat.isunlimited())
print(time.isunlimited())
````

When we have created the dimensions we are ready to create the actual variables. createVaribles takes 3 arguments the name of the variable, the datatype of the elements in the array and a tuple with the dimensions of the variable. 

```f8``` ```f4``` and ```i4``` are datatype descriptions of the individual elements in the arrays. f4 are 4 byte floating point, i4 4 byte integers etc. 

````python
times = rootgrp.createVariable('time','f8',('time',))
levels = rootgrp.createVariable('level','i4',('level',))
latitudes = rootgrp.createVariable('latitude','f4',('lat',))
longitudes = rootgrp.createVariable('longitude','f4',('lon',))
# two dimensions unlimited.
temp = rootgrp.createVariable('temp','f4',('time','level','lat','lon',))
temp.units = 'K'
````

````python
print rootgrp.variables['temp']
````

Now lets define our latitude and longitude coordinates and add them to the netCDF file. Note how we use the array index notation to add to the netCDF variables. 

````python
import numpy as np
lats =  np.arange(-90,91,2.5)
lons =  np.arange(-180,180,2.5)
nlats = len(lats)
nlons = len(lons)
latitudes[:] = lats
longitudes[:] = lons
````

We haven't stored any real data yet as can be seen by inspecting the shape of the ```temp``` variable. 

````python
rootgrp.variables['temp'].shape
````

Lets make up some data to insert into the ```temp``` variable.

````python
temp[0:5,0:10,:,:] = np.random.uniform(size=(5,10,nlats,nlons))
````

````python
rootgrp.variables['temp'].shape
````

We created 10 different levels when inserting the temperature data but we did not actually define the levels. Lets see how that looks. 

````python
levels.shape
````

````python
levels[:]
````

Now lets define some levels.

````python
levels[:] =  [1000.,850.,700.,500.,300.,250.,200.,150.,100.,50.]
````

````python
print(levels[:])
print(levels.dtype)
````

Remember how we made levels an array of integers. That is reflected by the fact that the floating point values that we added to levels are converted to integers.

To ensure that the data is written to disk correctly it is important that we close the file when we are done with adding data to it.

````python
rootgrp.close()
````

##Reading data from a multi-file netCDF dataset.

If you want to read data from a variable that spans multiple netCDF files, you can use the MFDataset class to read the data as if it were contained in a single file. Instead of using a single filename to create a Dataset instance, create a MFDataset instance with either a list of filenames, or a string with a wildcard (which is then converted to a sorted list of files using the python glob module). Variables in the list of files that share the same unlimited dimension are aggregated together, and can be sliced across multiple files. To illustrate this, let's first create a bunch of netCDF files with the same variable (with the same unlimited dimension). The files must in be in NETCDF3_64BIT, NETCDF3_CLASSIC or NETCDF4_CLASSIC format (NETCDF4 formatted multi-file datasets are not supported).

First we write some data split over a number of files:

````python
for nfile in range(10):
    with netCDF4.Dataset('mftest'+repr(nfile)+'.nc','w',format='NETCDF4_CLASSIC') as f:
        f.createDimension('x',None)
        x = f.createVariable('x','i',('x',))
        x[0:10] = np.arange(nfile*10,10*(nfile+1))
````

Above we use a cleaver Python trick. Rather than opening a file ```f``` and explicitly close it after reading from it we use the ```with``` context manager. This automatically closes the file when the context ends (that is when the indented section ends). Not only does that mean that we don't have to remember to close the file, which would be problematic here since we are opening several files in a loop, it also ensures that the file is closed even if an error happens while the file is open.

Now read all the files back in at once with MFDataset

````python
f = netCDF4.MFDataset('mftest*nc')
print f.variables['x'][:]
````
